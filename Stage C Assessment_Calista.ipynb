```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv('Data_for_UCI_named.csv')
df=dataset.copy()
```

```python
df
```

```python
df.describe()
```

```python
df.isnull().any()
```

```python
for x in df.index:
    if df.loc[x,'stab'] <= 0:
        df.loc[x,'stabf']='stable'
    else:
        df.loc[x,'stabf']='unstable'
```

```python
df=df.sample(frac=1)
df
```

```python
df=df.drop(['stab'],axis=1)
```

```python
df
```

```python
for x in df.index:
    if df.loc[x,'stabf'] =='stable':
        df.loc[x,'stabf']=1
    else:
        df.loc[x,'stabf']=0
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
x=df.drop(['stabf'],axis=1)
y=df['stabf']
```

```python
x.columns
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
type(ytrain)
```

```python
ytrain=np.array(ytrain)
ytrain=ytrain.reshape(-1,1)
ytrain.shape
```

```python
from sklearn.preprocessing import LabelEncoder
label_data = LabelEncoder()
ytrain = label_data.fit_transform(ytrain)
ytest=label_data.fit_transform(ytest)
```

```python
ytrain.shape
```

```python
Scaler=StandardScaler()
xtrain=Scaler.fit_transform(xtrain)
#ytrain=Scaler.fit_transform(ytrain)
xtest=Scaler.transform(xtest)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()                        #criterion='gini',max_depth=100,
                                                        #n_estimators=10,ccp_alpha= 0.01,random_state=30)
model.fit(xtrain,ytrain)
y_pred_rf=model.predict(xtest)
                                                         #ypred=model.predict(xtrain)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,log_loss
```

```python
cm_lr=confusion_matrix(ytest,y_pred_rf)
cm_lr
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,log_loss
```

```python
print(f'The Accuracy for the testing set is : ',model.score(xtest,ytest))
print(f'The Accuracy for the training set is : ',model.score(xtrain,ytrain))
```

```python
accuracy=accuracy_score(ytest,y_pred_rf.round())
precision=precision_score(ytest,y_pred_rf.round(),average='weighted')
recall=recall_score(ytest,y_pred_rf.round(),average='weighted')
f1score=f1_score(ytest,y_pred_rf.round(),average='weighted')

print('Accuracy:%.2f%%' %(accuracy*100.0))
print('Precision:%.2f%%' %(precision*100.0))
print('Recall:%.2f%%' %(recall*100.0))
print('F1-score:%.2f%%' %(f1score*100.0))
```
